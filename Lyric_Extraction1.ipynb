```python
# Libraries and Packages

!pip install git+https://github.com/openai/whisper.git
%pip install -U demucs
%pip install indic-transliteration
%pip install ai4bharat-transliteration
from indic_transliteration import sanscript
from indic_transliteration.sanscript import SchemeMap, SCHEMES, transliterate
from ai4bharat.transliteration import XlitEngine

# Function to transcribe the given audio file

import subprocess

def transcribe_audio(audio_file_path, model="small"):
    command = f"whisper \"{audio_file_path}\" --model {model}"

    try:
        result = subprocess.run(command, shell=True, capture_output=True, text=True)
        if result.returncode == 0:
            transcription = result.stdout.strip()
            return transcription
        else:
            error_message = result.stderr.strip()
            return f"Error: {error_message}"
    except Exception as e:
        return f"An error occurred: {str(e)}"

# Function to Transliterate the already transcribed audio file

def transliterate_tamil_to_english(text):
    # Set the source and target scripts
    target_language = "indic"
    src_lang = "ka"

    # Create a new XlitEngine instance
    e = XlitEngine(src_script_type=target_language, beam_width=10, rescore=False)
    out = e.translit_sentence(text, lang_code=src_lang)
    return out
```

```python
# Main function
def main():
    # Accept user input for audio path and option
    audio_path = input("Enter the path to the audio file: ")
    option = input("Enter '1' for Option-1 or '2' for Option-2: ")

    if option == '1':
        # Transcribing the Audio input given by the user
        transcription = transcribe_audio(audio_path)
        print("Transcription:")
        print(transcription)

    elif option == '2':
        # Transcribing the Audio input given by the user
        transcription = transcribe_audio(audio_path)
        # Read the transcribed content from a file
        with open('output.srt', 'r', encoding='utf-8') as file:
            transcribed_text = file.read().strip()

        # Transliterate the transcribed text to English
        english_text = transliterate_tamil_to_english(transcribed_text)
        print("Lyrics in English:")
        print(english_text)
    else:
        print("Invalid option. Please select '1' or '2'.")

if __name__ == "__main__":
    main()
```

```python
import whisper

model = whisper.load_model("base")
model.device

# load audio and pad/trim it to fit 30 seconds
audio_path = input("Enter the path to the audio file: ")
audio = whisper.load_audio(audio_path)
audio = whisper.pad_or_trim(audio)

# make log-Mel spectrogram and move to the same device as the model
mel = whisper.log_mel_spectrogram(audio).to(model.device)

# detect the spoken language
_, probs = model.detect_language(mel)
print(f"Detected language: {max(probs, key=probs.get)}")

# decode the audio
options = whisper.DecodingOptions()
result = whisper.decode(model, mel, options)

# print the recognized text
print(result.text)
```
