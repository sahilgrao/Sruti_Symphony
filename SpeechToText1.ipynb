```python
# Function to transcribe the given audio file

import subprocess

def transcribe_audio(audio_file_path, model="small"):
    command = f"whisper \"{audio_file_path}\" --model {model}"

    try:
        result = subprocess.run(command, shell=True, capture_output=True, text=True)
        if result.returncode == 0:
            transcription = result.stdout.strip()
            return transcription
        else:
            error_message = result.stderr.strip()
            return f"Error: {error_message}"
    except Exception as e:
        return f"An error occurred: {str(e)}"
```

```python
# Transcribing the Audio input given by the user

audio_path = input("Enter the path to the audio file: ")
transcription = transcribe_audio(audio_path)

print("Transcription:")
print(transcription)
```

```python
# Function to Transliterate the already transcribed audio file:
def transliterate_tamil_to_english(trans_path):
    # Set the source and target scripts
    with open(trans_path, 'r', encoding='utf-8') as file:
        input_text = file.read().strip()

    target_language = "indic"
    src_lang = "ta"

    # Create a new XlitEngine instance
    e = XlitEngine(src_script_type=target_language, beam_width=10, rescore=False)
    out = e.translit_sentence(input_text, lang_code=src_lang)
    return out

# Usage
trans_path = input("Enter the path to the transcribed file: ")
english_text = transliterate_tamil_to_english(trans_path)
print("Lyrics in English:")
print(english_text)
```
